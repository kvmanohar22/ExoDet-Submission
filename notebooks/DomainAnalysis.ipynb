```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

```python
data = np.load('../features/raw_mean_std_normalized.npy')
# data = load('../features/detrend_gaussian20.npy')
```

```python
index = 29
```

```python
plt.figure(figsize=(15,5))
plt.plot(data[index], 'o', markersize=3)
ax = plt.axes()
ax.set_xlim(left=0,right=3200)
ax.set_yticklabels([])
plt.title('Star %d with its spectogram, window size 32'%index)
plt.show()

plt.figure(figsize=(15,5))
f ,t, Sxx = signal.spectrogram(data[index], fs=1, nperseg=32)
ax = plt.axes()
ax.set_xlim(left=0, right=3200)
ax.set_yticklabels([])
plt.pcolormesh(t ,f, Sxx, cmap='BrBG_r')
plt.show()

# f, t, Zxx = signal.stft(data[index], fs=1, nperseg=64)
# plt.figure(figsize=(15,5))
# plt.pcolormesh(t ,f, np.abs(Zxx), cmap='BrBG_r')
# plt.title(cm)
# plt.show()

```
